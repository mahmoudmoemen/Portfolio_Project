```python
#Removing duplicates
df.drop_duplicates()
```

```python
# calculating number of missing values in each column
df.isnull().sum()
```

```python
#Removing unneeded columns
df.drop(columns=['ZIP CODE'],inplace=True)
df.drop(columns=['LOCATION'],inplace=True)
df.drop(columns=['CROSS STREET NAME'],inplace=True)
df.drop(columns=['OFF STREET NAME'],inplace=True)
df.drop(columns=['CONTRIBUTING FACTOR VEHICLE 2'],inplace=True)
df.drop(columns=['CONTRIBUTING FACTOR VEHICLE 3'],inplace=True)
df.drop(columns=['CONTRIBUTING FACTOR VEHICLE 4'],inplace=True)
df.drop(columns=['CONTRIBUTING FACTOR VEHICLE 5'],inplace=True)
df.drop(columns=['VEHICLE TYPE CODE 2'],inplace=True)
df.drop(columns=['VEHICLE TYPE CODE 3'],inplace=True)
df.drop(columns=['VEHICLE TYPE CODE 4'],inplace=True)
df.drop(columns=['VEHICLE TYPE CODE 5'],inplace=True)
```

```python
#converting all columns headers to lower case
df.columns= [x.lower() for x in df.columns]
df.columns
```

```python
#converting some columns types
df['crash date'] = df['crash date'].astype('datetime64')
df['crash time'] = df['crash time'].astype('datetime64')
```

```python
df
```

```python
#calculating the rate of accidents over the years
years =df.groupby(df['crash date'].dt.year)['collision_id'].count()
years.plot().set_title('Accidents rate over years',fontdict={'size': 15, 'weight': 'bold'})

plt.show()
```

```python
# Find out the most months in which accidents occur throughout the year
accident_month = df.groupby(df['crash date'].dt.month)['collision_id'].count()
accident_month.plot().set_title('Accident rate over months',fontdict={'size': 15, 'weight': 'bold'})

plt.show()
```

```python
#discover accident rate over day hours
hourly_accident = df.groupby(df['crash time'].dt.hour)['collision_id'].count()
hourly_accident.plot().set_title('Accident rate over day hours',fontdict={'size': 15, 'weight': 'bold'})

plt.show()
```

```python
# calculating the ten contributing factor vehicle
contributing_factor = df.groupby(df['contributing factor vehicle 1'])['collision_id'].count()
factor =contributing_factor.sort_values(ascending = False).nlargest(10)
factor.plot.barh().set_title('The ten most  contributing factor vehicle',fontdict={'size': 15, 'weight': 'bold'})
plt.show()
```

```python
# calculating the ten most accidents streets
accidents_streets = df.groupby(df['on street name'])['collision_id'].count()
factor =accidents_streets.sort_values(ascending = False).nlargest(10)
factor.plot.barh().set_title('The ten most  accidents streets',fontdict={'size': 15, 'weight': 'bold'})
plt.show()
```

```python
#calculating injured rate over years
injured_people = df.groupby(df['crash date'].dt.year)['number of persons injured'].sum()
injured_people.plot().set(title ='Injured rate over years')
plt.show()
```

```python
#number of injured in each category
pedestrians_injured = df['number of pedestrians injured'].sum()
motorist_injured = df['number of motorist injured'].sum()
cyclist_injured = df['number of cyclist injured'].sum()
injured = [pedestrians_injured , motorist_injured , cyclist_injured]
x_axis = ['pedestrians_injured' , 'motorist_injured' , 'cyclist_injured']
plt.bar (x_axis , injured)
```

Know the rate of killed people over the years

```python
# calculatig killed rate over years
killed_people = df.groupby(df['crash date'].dt.year)['number of persons killed'].sum()
killed_people.plot().set(title='Killed rate over years')
plt.show()
```

```python
#calculating number of killed people in each category
pedestrians_killed = df['number of pedestrians killed'].sum()
motorist_killed = df['number of motorist killed'].sum()
cyclist_killed = df['number of cyclist killed'].sum()
x_axis = ['pedestrians_killed' , 'motorist_killed' , 'cyclist_killed']
killed = [pedestrians_killed , motorist_killed , cyclist_killed]
plt.bar (x_axis , killed)
```
